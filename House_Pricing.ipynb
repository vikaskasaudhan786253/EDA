### removing outliers

```python
q1 = data["living area"].quantile(0.25)
q1
```

```python
q3 = data["living area"].quantile(0.75)
q3
```

```python
iqr = q3-q1
```

```python
min_r = q1 - (1.5*iqr)
max_r = q3 + (1.5*iqr)
```

```python
min_r,max_r
```

```python
data = data[data["living area"]<=max_r]
```

```python
sns.boxplot(x="living area",data=data)
plt.show()
```
